### Criação de uma função para detalhamento dos dados Null ( data null ) e Types


```python
# Função para mapear dos dados do dataset 

def summary_missing_values(data):
    missing_values = pd.DataFrame(df_raw.isna().sum()/df_raw.shape[0])
    missing_values.reset_index(drop=True, inplace=True)    
    missing_values.columns = ["Absoluto dados ausentes(%)"]
    missing_values["Column"] = pd.Series(df_raw.columns).values
    # Make index be the column "Absolute Missing (%)"
    missing_values.reset_index(drop=True, inplace=True)
    missing_values.set_index("Column", inplace=True)
    
    # Calculando os valores comparando com o tatal de dados por coluna e os dados ausentes quantitativos em percentual
    column_missing = []
    column_total   = []
    column_miss    = []
    column_type    = []
    
    for column in df_raw.columns:
        col_miss = df_raw[column].isnull().sum()
        total_values = df_raw[column].count()       
        column_miss.append(col_miss)
        column_total.append(total_values) 
        column_missing.append((col_miss/total_values))
        column_type.append(column_type)
        
    # Construindo o dataframe que receberá os valores
    missing_values["Perc. Abs. ausentes(%)"] = pd.Series(column_missing).values
    missing_values["Perc. da equivalência(%)"] = -(pd.Series(column_missing).values-1)
    missing_values["Vlres únicos"] = pd.Series(column_total).values
    missing_values["Valores abs ausentes"] = pd.Series(column_miss).values
    missing_values['Variáveis tipos'] = pd.DataFrame(df_raw.dtypes, columns=['column_type'])
    missing_values.sort_values(by="Absoluto dados ausentes(%)", ascending=False, inplace=True)
    missing_columns = list(missing_values.query("`Absoluto dados ausentes(%)` > 0").index)
    
    pd.options.display.float_format = '{:.2%}'.format
    return missing_values, missing_columns

print('\n\t\t\t\t\t\tMATRIZ DE MAPEAMENTO DOS DADOS\n'.center(100))

missing_values, missing_columns_names = summary_missing_values(df_raw)
display(missing_values)
```

```python
def statistic_summary(df_raw): 
    df_raw = df_raw[numerical_features]   
    # count_ = pd.DataFrame(df_raw.apply(np.size))
    # pct_ = pd.DataFrame(df_raw.apply(np.size)/len(numerical_features)*100)
    
    sum_ = pd.DataFrame(df_raw.apply(np.sum))
    mean_ = pd.DataFrame(df_raw.apply(np.mean))
    std_ = pd.DataFrame(df_raw.apply(np.std))
    median_ = pd.DataFrame(df_raw.apply(np.median))
    
    min_ = pd.DataFrame(df_raw.apply(min))
    max_ = pd.DataFrame(df_raw.apply(max))
    
    range_ = pd.DataFrame(df_raw.apply(lambda x: x.max() - x.min()))
    q1_ = pd.DataFrame(df_raw.apply(lambda x: np.percentile(x,25)))
    q3_ = pd.DataFrame(df_raw.apply(lambda x: np.percentile(x,75)))
    
    skewness_= pd.DataFrame(df_raw.apply(lambda x: x.skew()))
    kurtosis_ = pd.DataFrame(df_raw.apply(lambda x: x.kurtosis()))
    
    # summary = pd.concat([count_, pct_, mean_, std_, median_, sum_, min_, max_, range_, q1_,q3_],axis=1)
    summary = pd.concat([mean_, std_, median_, sum_, min_, max_, range_, q1_, q3_, skewness_, kurtosis_ ],axis=1)
    summary.columns = ['MEDIA','DESVIO','MEDIANA','SOMA','MIN','MAX', 'RANGE', '25%','75%', 'SKEW', 'KURTOSIS']

    cm = sns.light_palette('Blue', as_cmap = True)
    display(summary.style.format('{:.2f}').background_gradient(subset=(summary.index[2:],summary.columns[:]),cmap=cm))

    # Gerando um relatório do dataset concatenado para ser consumido em um describe diferente do tradicional

print('\n\tDESCRITIVO ESTATÍSTICO COM MÉTODOS SKEW E KURTOSIS DAS MÉTRICAS DE DISPERSÃO IQR E CENTRAIS\n'.center(150))

statistic_summary(df_raw)
```

### Criação de uma função para plotagens dos gráficos e cenarizações.
  - Para cada plot, gráfico criado, chamo essa function para tratar do refinamento dos gráficos
   - Parâmetros plt %matplotlib para as labels 
   - Bar: gráficos verticais
   - Barh: gráficos horizontais
___

```python
### Function 
'''
  Dbian Data Scientist 
  Função responsável por anotar os indicadores das barras no gráfico.
  Entrada: tipo do gráfico ('bar' ou 'barh')
'''
def autolabel(tipo):
  if tipo == 'bar':
    for bar in plot.patches:
      plot.annotate(format(bar.get_height()),
                xy = (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                xytext=(0, 8),
                textcoords='offset points',
                ha='center', 
                va='center',
                size=11)
  elif tipo == 'barh':
    for bar in plot.patches:
      plot.annotate(format(bar.get_width()),
                  xy = (bar.get_width() + 0.3, bar.get_y() + 0.38),
                  xytext=(15, -2),
                  textcoords='offset points',
                  ha='center', 
                  va='center',
                  size=11)
    
# Função para retirar o contorno dos plots - gráficos. 
def gca():
    for posicao in ['right', 'top', 'bottom', 'left']: 
        plt.gca().spines[posicao].set_visible(False) 
    return None
```
